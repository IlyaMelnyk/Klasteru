```python
# Importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Importing the dataset
df = pd.read_csv('hyundi.csv',sep=',')
```

```python
# Drop useless
df = df.drop(['model'], axis='columns', inplace=False)
```

```python
# Cheking Missing data
df.isnull().sum()
```

пропущенных значений нет

```python
from sklearn import preprocessing
dfm = df
dfm['transmission'] = preprocessing.LabelEncoder().fit_transform(dfm['transmission'])
dfm['fuelType'] = preprocessing.LabelEncoder().fit_transform(dfm['fuelType'])
dfm
```

```python
import seaborn as sns
sns.set()
sns.pairplot(df)
```

```python
df.columns
```

```python
#renaming columns
df.columns = ['year', 'price', 'transmission', 'mileage', 'fuelType', 'tax', 'mpg', 'engineSize']
```

```python
# Dataset Statistics
df.describe()
```

```python
dfm = df
# Outliers
year_ = []
for year in df['year']:
    if year > (df['year'].mean() + 3*df['year'].std()):
        year = (dfm['year'].mean() + 3*df['year'].std())
    elif year < (df['year'].mean() - 3*df['year'].std()):
        year = (dfm['year'].mean() - 3*df['year'].std())
    year_.append(year)
df['year'] = year_

mpg_ = []
for mpg in df['mpg']:
    if mpg > (df['mpg'].mean() + 3*df['mpg'].std()):
        mpg = (dfm['mpg'].mean() + 3*df['mpg'].std())
    elif mpg < (df['mpg'].mean() - 3*df['mpg'].std()):
        mpg = (dfm['mpg'].mean() - 3*df['mpg'].std())
    mpg_.append(mpg)
df['mpg'] = mpg_

price_ = []
for price in df['price']:
    if price > (df['price'].mean() + 3*df['price'].std()):
        price = (dfm['price'].mean() + 3*df['price'].std())
    price_.append(price)
df['price'] = price_

tax_ = []
for tax in df['tax']:
    if tax > (df['tax'].mean() + 3*df['tax'].std()):
        tax = (dfm['tax'].mean() + 3*df['tax'].std())
    tax_.append(tax)
df['tax'] = tax_
```

```python
# Dataset Statistics
df.describe()
```

```python
# Feature Scaling
from sklearn.preprocessing import MinMaxScaler
sc = MinMaxScaler().fit(df)
df = sc.transform(df)
df = pd.DataFrame(df, columns = ['year', 'price', 'transmission', 'mileage', 'fuelType', 'tax', 'mpg', 'engineSize']).round(2)
```

```python
# Cheking correlations
import seaborn as sns
corrmat = df.corr()
f, ax = plt.subplots(figsize=(15, 12))
sns.heatmap(corrmat, annot=True)
plt.show()
```

# K-Means Clustering

```python
# Using the elbow method to find the optimal number of clusters
from sklearn.cluster import KMeans
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 0)
    kmeans.fit(df)
    wcss.append(kmeans.inertia_)
plt.plot(range(1, 11), wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
# Using the silhouette method to find the optimal number of clusters
from sklearn.metrics import silhouette_score

sil = []

for k in range(2, 7):
  kmeans = KMeans(n_clusters = k).fit(df)  
  preds = kmeans.fit_predict(df)
  sil.append(silhouette_score(df, preds, metric = 'euclidean'))


plt.plot(range(2, 7), sil)
plt.title('Silhouette Method')
plt.xlabel('Number of clusters')
plt.ylabel('Sil')
plt.show()

for i in range(len(sil)):
 print(str(i+2) +":"+ str(sil[i]))
```

в данном методе также рекомендуется выделить 2 кластера, но для большей информативности мы выделим 6.

```python
# Fitting K-Means to the dataset
km = KMeans(n_clusters = 6, init = 'k-means++', random_state = 0).fit_predict(df)
kms = KMeans(n_clusters = 6, random_state = 0).fit(df)
pd.DataFrame(kms.cluster_centers_, columns = ['year', 'price', 'transmission', 'mileage', 'fuelType', 'tax', 'mpg', 'engineSize']).round(2)
```

#### В первом кластере находятся машины, которые выпущены недавно, относительно небольшой стоимости с автомат. коробкой передач, бензин. топливом и небольшим пробегом.
#### Во втором кластере похожие машины, но стоят они дороже и работают на дизельном топливе.
#### В третьем: автомобили постарше с низкой стоимостью, налоговой ставкой и малолитражном двигателе, работающие на бензин. топливе.
#### В четвертом: новые машины высокой стоимости с механич. коробкой передач, небольшим пробегом, дизельным топливом и большим обьемом двигателя.
#### Пятый кластер похож на третий, но там машины немного дороже и ездят на дизельном топливе; у них также большой обьем двигателя и высокий расход mpg (miles per gallon).
#### Наконец в шестом кластере находятся самые новые автомобили с высокой ценой, механич. коробкой передач, на бензин. топливе, с высокой налог. ставкой, но низким расходом топлива.
